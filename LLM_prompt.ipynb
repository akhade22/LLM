```python
from transformers import TFAutoModelForSeq2SeqLM, AutoTokenizer
```

```python
CheckpointMoodel= "google/flan-t5-large"
```

```python
#CheckpointMoodel="google/flan-t5-large"
tokenizer= AutoTokenizer.from_pretrained("google/flan-t5-large" )
model= TFAutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large" )
```

```python
prom="King Kakudmi, ruler of Kusasthali (modern-day Dwarka), had a beautiful and virtuous daughter, Revati. Wanting the best husband for her, he decided to personally seek advice from Lord Brahma, the creator of the universe. Kakudmi and Revati traveled to Brahmaloka, the celestial abode of Lord Brahma, where they waited patiently for a chance to speak with him. However, time moves differently in different realms. A short time in Brahmaloka was equivalent to thousands of years on Earth. When Brahma finally addressed them, he laughed and informed Kakudmi that time had passed so differently that many generations had come and gone on Earth. All the suitors Kakudmi had considered for Revati were long dead, and civilizations had changed. Realizing their predicament, Brahma advised Kakudmi to seek Lord Balarama, the brother of Krishna, as a suitable husband for Revati. They returned to Earth, only to find it completely different from what they had known. Since people had become much shorter than those in their time, Revati, who was extraordinarily tall, had to be adjusted—Lord Balarama lightly pressed her shoulder with his plough weapon to bring her to the right size. They were then married. Who was King Kakudmi?"
```

```python
input=tokenizer(prom,return_tensors="tf",max_length=512,truncation=True,padding=True)
```

```python
outout=model.generate(**input,max_new_tokens=512,min_new_tokens=10,do_sample=True)
```
